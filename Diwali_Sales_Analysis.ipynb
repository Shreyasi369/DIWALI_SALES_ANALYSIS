```python
# Values Countplot 
a =sns.countplot (x = 'Gender', data = df)
for bars in a. containers:
    a.bar_label(bars)
```

```python
# GroupBy for Gender
People= df.groupby(['Gender'],as_index= False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)
sns.barplot (x = 'Gender', y= 'Amount',data = People)
```

```python
df.columns
```

```python
# Age group count by bar plots 
a = sns.countplot(data = df,x = 'Age Group',hue= 'Gender')
for bars in a.containers:
    a.bar_label(bars)
```

```python
# GroupBy for Age group
People = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending= False)
sns. barplot (x= 'Age Group', y= 'Amount',data= People)
```

```python
# total numbers of orders from all state
Sales_state= df. groupby(['State'],as_index=False)['Orders'].sum().sort_values(by = 'Orders', ascending=False).head(10)
sns.set(rc= {'figure.figsize':(25,5)})
sns.barplot(data= Sales_state,x = 'State',y='Orders') 
```

```python
# total numbers of Amount from all state
Salse_Amount= df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by = 'Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(data= Salse_Amount,x= 'State',y = 'Amount')
```

```python
# count of Merrital_Status
a= sns.countplot(data=df, x= 'Marital_Status')
sns.set(rc={'figure.figsize':(15,8)})
for bars in a.containers:
    a.bar_label(bars)
```

```python
# numbers of Amount for Marital_Status
Salse_Amount = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by= 'Amount',ascending=False)
sns.set({'figure.figsize':(10,3)})
sns.barplot(data= Salse_Amount,x= 'Marital_Status',y= 'Amount',hue= 'Gender')
```

```python
# Count by Occupation
a = sns.countplot (data= df,x = 'Occupation')
sns.set(rc= {'figure.figsize':(20,5)})
for bars in a. containers:
    a.bar_label(bars)
```

```python
# numbers of Amount for Occupation
Salse_Amount = df.groupby(['Occupation','Gender'],as_index=False)['Amount'].sum().sort_values(by= 'Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=Salse_Amount,x='Occupation',y='Amount',hue= 'Gender')
```

```python
#count of Product category
a = sns.countplot(data=df,x = 'Product_Category')
sns.set(rc={'figure.figsize':(20,10)})
for bars in a.containers:
    a.bar_label(bars)
```

```python
Salse_Amount = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by= 'Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(data=Salse_Amount,x='Product_Category',y= 'Amount')
```

```python
# top Saleing Prouct_ID
Salse_Pro= df.groupby(['Product_ID'],as_index=False)['Amount'].sum().sort_values(by= 'Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=Salse_Pro,x='Product_ID',y='Amount')
```
